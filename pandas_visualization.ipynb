```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
sns.set(color_codes=True) # You need this code to convert matplot color to seaborn color
```

```python
df1 = pd.read_csv("data/df1")
df1.head(5)
```

```python
df1 = pd.read_csv("data/df1", index_col=0)
df1.head(5)
```

```python
df2 = pd.read_csv("data/df2")
df2.head(5)
```

```python
df1['A'].hist() 
# histogram of A column from df1
# It is calling matplotlib under the hood
```

## Historam using seaborn

```python
sns.distplot(df1["A"]) #seaborn
```

## seaborn reference
* https://seaborn.pydata.org/tutorial/distributions.html

```python
sns.distplot(df1["A"], kde= False) # KDE stands for kernel density estimate
```

```python
df1['A'].hist(bins=20) 
# increase the number of bins
```

```python
df1['A'].plot(kind='hist', bins=30) # All plots can be done through this command. pandas has several plot types
```

```python
df1['A'].plot.hist(bins=30)  # call hist on plot
```

## Area plot

```python
df2.plot.area(alpha=0.6) # alpha sets the transparency
```

## Bar chart

```python
df2.plot.bar(alpha=0.8, figsize=(12,4))
```

```python
df2.plot.bar(alpha=0.8, stacked=True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
df1['A'].plot.line(x=df1.index, y='B',figsize=(12,5), lw=1)  # In the line plot, we have to specify X and Y
```

## Scatter plot

```python
df1.plot.scatter(x='A', y='B',figsize=(10,7))
```

```python
df1.plot.scatter(x='A', y='B', c='C', figsize=(10,7), cmap='coolwarm')
# You can show three level of information in 2d plot by color "c="
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*125, figsize=(10,7))
# 3d dimension by size "s="
# You have to pass "s=" the dataframe
# multiply the s by an appropriate scale
```

## Seaborn joinplot

```python
sns.jointplot(x="A", y="B", data=df1, height=7) # This is a joint plot
```

## Box plot

```python
df2.plot.box(figsize=(10,7))
# max, 75%, median, 25%, min
```

```python
df = pd.DataFrame(np.random.randn(300,2), columns=['a','b'])
df.head()

# np.random.randn(300,2) creates a 300 x 2 array of standard normal distribution
# "columns =" sets the name of the columns  
```

## Hexagonal Binning

```python
df.plot.hexbin(x='a', y='b', gridsize=20)
```

## Kernel density estimation: KDE

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density() # density generates the same thing
```

```python
df2.plot.density() # put all KDE in the same plot
```
