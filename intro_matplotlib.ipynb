## Matplotlib Basics
* Reference : https://www.udemy.com/python-for-finance-and-trading-algorithms/learn/v4/overview

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11) # create a linear space
y = x**2
```

## Functional method

```python
plt.plot(x,y) #functional approach

# plt.show() # plt.show() is printing the plot. You don't have to use this in jupyter
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

## Multiple plots on the same campus

```python
plt.subplot(1,2,1) # row, columns, plot number
plt.plot(x, y, "r")

plt.subplot(1,2,2)
plt.plot(y,x, "b")
```

## Object oriented method

```python
fig = plt.figure() # Create a figure object. This is a blank campus
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
# Call the figure object, add_axes method, reassign it to axes. 
# left margin, bottom margin, width, and height of the campus
axes.set_xlabel("X lablel")
axes.set_ylabel("Y lablel")
axes.set_title("Set Title")
axes.plot(x,y)
```

```python
fig = plt.figure() # Create a figure object
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left margin, bottom margin, width, and height of the campus
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure() # Create a figure object
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left margin, bottom margin, width, and height of the campus
axes2 = fig.add_axes([0.4, 0.2, 0.4, 0.3])
```

```python
fig = plt.figure() # Create a figure object manually
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
# Manually add axes to the fig 
# left margin, bottom margin, width, and height of the campus
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.set_title("larger plot")
axes2.set_title("smaller plot")
axes1.plot(x,y)
axes2.plot(y,x)
```

## plt.subplots()

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes # axes is a list of matplotlib axis
```

```python
# plt.subplots() do fig.add_axes() stuff automatically based on how many rows and columns you want
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
plt.tight_layout() # This will take care of overlapping plot
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")
plt.tight_layout() # This will take care of overlapping plot
```

## Figure Size and DPI

```python
fig = plt.figure(figsize = (3,2)) # First argument is figsize, second is dpi. Pass width and height as as tuple
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize = (8,2)) # 8 inches by 2 inches
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(figsize = (8,2))
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

## Legends, labels and titles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared') #set label argument if there are multiple axes
ax.plot(x,x**3, label = 'X Cubed')
ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')

ax.legend(loc=0) 
# 'loc = 0' is the location code.
# See legend doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared') #set label argument if there are multiple axes
ax.plot(x,x**3, label = 'X Cubed')
ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')

ax.legend(loc = 1) # see legend doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label ='X Squared') #set label argument if there are multiple axes
ax.plot(x,x**3, label = 'X Cubed')
ax.set_title('Title')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')

ax.legend(loc = (0.05,0.85)) 
# you can set a specific location
# see legend doc: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
```

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 120])
axes[2].set_xlim([0, 5])
axes[2].set_title("custom axes range");
```
